```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
n1 = np.arange(1,10)
```

```python
n1
```

```python
y = n1*3
```

```python
y
```

```python
plt.plot(n1,y,color="g",linestyle=":",linewidth=4)

plt.title("graph of x and Y")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

```python
x = np.arange(1,11)
```

```python
y1 = x*2
y2 = x*3
```

```python
plt.plot(x,y1,color="g",linestyle="-",linewidth=5)
plt.plot(x,y2,color="r",linewidth=3)
plt.title("The graph of Rahul")
plt.xlabel("This side represent x side")
plt.ylabel("This side represent y side")
plt.grid(True)
plt.show()
```

```python
# Adding subplots
```

```python
x = np.arange(1,11)
```

```python
y = x*3
```

```python
z = x*4
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color="g",linestyle=":",linewidth=2)
plt.title("The graph of x,y")
plt.xlabel("This side represent x axis")
plt.ylabel("This side represent y axis")
plt.grid(True)







plt.subplot(1,2,2)
plt.plot(x,z,color="r",linestyle=":",linewidth=2)
plt.title("This graph represent the x and z")
plt.xlabel("This side is x axis ")
plt.ylabel("This side is y label")
plt.grid(True)
```
